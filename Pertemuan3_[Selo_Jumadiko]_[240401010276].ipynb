```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
import requests
```

```python
# step 0 - Load & eksplorasi awal
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1LfQWProB0VjWN5q8bKuRIgn-stULfIRo')
print('Shape awal :', df.shape)
print(df.isnull().sum())

# Setelah di eksplorasi terdapat 130 baris dan 7 kolom
# kemudian terdapat nilai yang hilang di kategori luas_m2 18 nilai, harga_juta 17 nilai, dan kamr 10 nilai
```

```python
# langkah 1 - Hapus duplikasi
df.drop_duplicates(inplace=True)
print('Shape setelah dihapus duplikasi :', df.shape)
```

```python
# langkah 2 - Normalisasi String
df['kota'] = df['kota'].astype(str).str.strip().str.title()
df['kondisi'] =df['kondisi'].astype(str).str.strip().str.lower()
```

```python
# langkah 3 - Imputasi Missing Values
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

```python
# langkah 4 - Tangani Outlier (IQR Fence)
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
  Q1, Q3 = df[col].quantile([0.25, 0.75])
  IQR = Q3 - Q1
  df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)
```

```python
# langkah 5 — Validasi & Ekspor
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```
