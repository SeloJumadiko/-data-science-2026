```python
import pandas as pd
import numpy as np

# Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

print(df.shape, df.columns.tolist())

# Q1 Berapa jumlah baris dan kolom dataset Titanic?
# A1 terdapat 891 baris dan 12 kolom
```

```python
#Kolom memiliki nilai kosong
print(df.isnull().sum())

# Q2 Kolom mana yang memiliki nilai kosong (missing)?
# A2 Terdapat 3 kolom yang kosong yaitu kolom Age, Cabin & Embarked
```

```python
# Presentase penumpang selamat
df['Survived'].head()
jumlah_selamat = df['Survived'].sum()
total_penumpang = len(df)
presentase_selamat = (jumlah_selamat / total_penumpang) * 100
print(presentase_selamat)

# Q3 Berapa persen penumpang yang selamat ?
# A3 Presentase penumpang yang selamat adalah 38.4 %
```

```python
# Rata-rata usia penumpang wanita kelas 1 yang selamat

wanita_kelas_1 = df[
    (df['Sex'] == 'female') &
     (df['Pclass'] == 1) &
    (df['Survived'] == 1)]
({"jumlah penumpang": [len (wanita_kelas_1)],
  "rata-rata usia": [wanita_kelas_1['Age'].mean()]
  } )

# Q4 Berapa rata-rata usia penumpang wanita kelas 1 yang selamat
# A4 Rata-rata usia penumpang wanita kelas 1 yang selamat adalah 35 tahun
```

```python
# Kelas penumpang yang tingkat keselamatan tertinggi?

survival_rate = df.groupby('Pclass')['Survived'].mean() * 100
kelas_tertinggi = survival_rate.idxmax()
tingkat_keselamatan_tertinggi = survival_rate.max()
print("survival rate per kelas:")
print(survival_rate)

# Q5 Kelas penumpang mana yang tingkat keselamatan tertinggi?
# A5 Kelas penumpang yan memiliki keselamatan tertinggi adalah kelas 1
```

```python
# Tampilkan hasil groupby('Pclass')['Survived'].mean()
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```
